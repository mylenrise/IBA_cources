3) Is it better for student's academic performance if his/her parents work at the same job ('other' should be excluded)?

```python
res=df[(df['Mjob'] == df['Fjob']) & (df['Mjob'] != 'other')]

print("Avegare annual academic performance if having both parents at the same job: ",(res.G1.mean()+res.G2.mean()+res.G3.mean())/3)
print("Avegare annual academic performance among all students:",(df.G1.mean()+df.G2.mean()+df.G3.mean())/3)

res2 = df.groupby(by = ['Mjob', 'Fjob'])['G3'].mean()

print(res2)

res3 = df.pivot_table(
    values = 'G3', 
    columns = 'Mjob', 
    index = 'Fjob', 
    aggfunc = 'mean'
)


res3
```

As we can see, it's better for student's academic performance if his/her parents work at the same job

4) Who is more healthy: students with final grade not less than 15 or not?

```python
good_grades=df[(df['G3'] >= 15)]
bad_grades=df[(df['G3'] < 15)]

print("Students with final grade not less than 15 health average mark: ", good_grades.health.mean())
print("Students with final grade less than 15 health average mark: ", bad_grades.health.mean())


res2 = df.groupby(by = 'G3')['health'].mean()

print(res2)

res3 = df.pivot_table(
    values = 'health', 
    index = 'G3',  
    aggfunc = 'mean'
)

print(res2[15:].count())
print(res2[:15].count())
res3
```

We can make an interesting conclusion: students with final grade not less than 15 are more healthy if we count mean health index amount all students in the group. But if we pay attention on statistics of health of each mark we'll see that those who have mark 7,12,9 have the best average health among students with similar marks.

5) Is there a direct link between going out with friends and final grade?

```python
goout = []
for i in range(1,6):
    goout.append(df[df['goout']==i])
    print ('Going out idnx:',i,', mean final grade',goout[i-1].G3.mean())
    
res2 = df.groupby(by = 'goout')['G3'].mean()

print(res2)

res3 = df.pivot_table(
    values = 'G3', 
    index = 'goout',  
    aggfunc = 'mean'
)
res3
```

6) Is the a connection between Mother's education and desire to take higher education?

```python
medu = []
for i in range(1,5):
    medu.append(df[df['Medu']==i])
    print ('Medu idnx:',i,', wish to take higher education:',medu[i-1]['higher'].value_counts())

    
res2 = df.groupby(by = ['higher'])['Medu'].mean()

print(res2)

res3 = df.pivot_table(
    values = 'Medu', 
    index = 'higher',  
    aggfunc = 'mean'
)
```

7) Could we say that girls have more failures when they have romantic relationship than boys?

```python
print(df.groupby(by=['sex', 'romantic','failures'])['failures'].count())

df1 = df[(df['sex'] == 'F') & (df['failures'] > 0) & (df['romantic'] == 'yes')]
df2 = df[(df['sex'] == 'M') & (df['failures'] > 0) & (df['romantic'] == 'yes')]

p_table1 = df1.pivot_table(
    values = 'failures', 
    columns = 'sex', 
    index = 'romantic', 
    aggfunc = 'count'
)

p_table2 = df2.pivot_table(
    values = 'failures', 
    columns = 'sex', 
    index = 'romantic', 
    aggfunc = 'count'
)

print(p_table1)
print(p_table2)
```

As we can see, girls have more failures when they have romantic relationship

12) Calculate average final grade by quality of family relationship and parent's cohabitation status. Which case has the biggest average grade?

```python
famrelA = []
famrelT = []
for i in range(1,6):
    famrelA.append(df[(df['famrel']==i) & (df['Pstatus']=='A')])
    famrelT.append(df[(df['famrel']==i) & (df['Pstatus']=='T')])
    print ('Famrel idnx:',i,',Pstatus A, mean final grade',famrelA[i-1].G3.mean())
    print ('Famrel idnx:',i,',Pstatus T, mean final grade',famrelT[i-1].G3.mean())

res2 = df.groupby(by = ['Pstatus','famrel'])['G3'].mean()

print(res2)

res3 = df.pivot_table(
    values = 'G3',
    columns = 'famrel',
    index = 'Pstatus',  
    aggfunc = 'mean'
)
res3
```

As we can see, students with Pstatus A in average have better final grades. Students with Pstatus A and famrel index 5 have the biggest average grade
